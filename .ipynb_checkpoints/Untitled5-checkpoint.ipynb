Load our training and test data frames into main memory. 
FullVisitorId column converted to string format.
4 columns (device, geoNetwork, totals, trafficSource) are in JSON format. So we normalize those columns.

```python
def load_df(csv_path='train.csv', nrows=None):
    JSON_COLUMNS = ['device', 'geoNetwork', 'totals', 'trafficSource']
    
    df = pd.read_csv(csv_path, 
                     converters={column: json.loads for column in JSON_COLUMNS}, 
                     dtype={'fullVisitorId': str},
                     nrows=nrows)
    
    for column in JSON_COLUMNS:
        column_as_df = json_normalize(df[column])
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns]
        df = df.drop(column, axis=1).merge(column_as_df, right_index=True, left_index=True)
    print(f"Loaded {os.path.basename(csv_path)}. Shape: {df.shape}")
    return df
```

```python
def ImputeMissingValues(train,test):
    for df in [train, test]:
        df['trafficSource.adContent'].fillna('N/A', inplace=True)
        df['trafficSource.adwordsClickInfo.slot'].fillna('N/A', inplace=True)
        df['trafficSource.adwordsClickInfo.page'].fillna(0.0, inplace=True)
        df['trafficSource.adwordsClickInfo.isVideoAd'].fillna('N/A', inplace=True)
        df['trafficSource.adwordsClickInfo.adNetworkType'].fillna('N/A', inplace=True)
        df['trafficSource.adwordsClickInfo.gclId'].fillna('N/A', inplace=True)
        df['trafficSource.isTrueDirect'].fillna('N/A', inplace=True)
        df['trafficSource.referralPath'].fillna('N/A', inplace=True)
        df['trafficSource.keyword'].fillna('N/A', inplace=True)
        df['totals.bounces'].fillna(0.0, inplace=True)
        df['totals.newVisits'].fillna(0.0, inplace=True)
        df['totals.pageviews'].fillna(0.0, inplace=True)
```

<font size=8>Task 5 : Addition of External Data Set </font>

I have taken external data set from https://www.kaggle.com/satian/exported-google-analytics-data. There are 4 files : Test_external_data.csv, Test_external_data_2.csv, Train_external_data.csv, Train_external_data_2.csv.
They contain 6 columns : Client Id, Sessions, Avg_Session_Duration, Bounce_Rate, Revenue Transactions, Goal Conversion Rate. It really helps in making the accurate predictions of the total revenue per user. My rank improves from 846 to 319 after incorporating this data set into my training models.

Let's work on external data set. I merged the columns of external data set into our existing data sets.

```python
train_data = pd.read_csv('Train_external_data.csv', low_memory=False, skiprows=6, dtype={"Client Id":'str'})
test_data = pd.read_csv('Test_external_data.csv', low_memory=False, skiprows=6, dtype={"Client Id":'str'})
for df in [train_data,test_data]:
    df["visitId"] = df["Client Id"].apply(lambda x: x.split('.', 1)[1]).astype(str)  
train = train.merge(train_data, how="left", on="visitId")     
test = test.merge(test_data, how="left", on="visitId")
for df in [train, test]:
    df.drop("Client Id", 1, inplace=True)
    df["Revenue"].fillna('$', inplace=True)
    df["Revenue"] = df["Revenue"].apply(lambda x: x.replace('$', '').replace(',', ''))
    df["Revenue"] = pd.to_numeric(df["Revenue"], errors="coerce")
    df["Revenue"].fillna(0.0, inplace=True)
    df["Sessions"] = df["Sessions"].fillna(0)
    df["Avg. Session Duration"] = df["Avg. Session Duration"].fillna(0)
    df["Bounce Rate"] = df["Bounce Rate"].fillna(0)
    df["Revenue"] = df["Revenue"].fillna(0)
    df["Transactions"] = df["Transactions"].fillna(0)
    df["Goal Conversion Rate"] = df["Goal Conversion Rate"].fillna(0)
    df['date_new'] = pd.to_datetime(df['visitStartTime'], unit='s')
    df['date'] = df['date_new'].dt.date
    df['sess_date_dow'] = df['date_new'].dt.dayofweek
    df['sess_date_hours'] = df['date_new'].dt.hour
    df['sess_month'] = df['date_new'].dt.month
```

Approach : I did model training with GroupKfold Validation in LightGBM Regressor. I first trained model on session level and then on visitor level. The performance could further be improved using stack of different models. As of now I used only a single model, i.e LightGBM Regressor.

```python
# K-Fold Validation
def get_folds(df=None, n_splits=5):
    unique = np.array(sorted(df['fullVisitorId'].unique()))
    folds = GroupKFold(n_splits=n_splits)
    fold_ids = []
    ids = np.arange(df.shape[0])
    for train, val in folds.split(X=unique, y=unique, groups=unique):
        fold_ids.append(
            [
                ids[df['fullVisitorId'].isin(unique[train])],
                ids[df['fullVisitorId'].isin(unique[val])]
            ]
        )
    return fold_ids
```

```python
y_reg = train['totals.transactionRevenue'].fillna(0)
del train['totals.transactionRevenue']

excluded_features = [
    'visitId', 'visitStartTime','date', 'date_new', 'fullVisitorId', 'sessionId', 'totals.transactionRevenue']
categorical_features = [
    _f for _f in train.columns
    if (_f not in excluded_features) & (train[_f].dtype == 'object')]

for f in categorical_features:
    train[f], indexer = pd.factorize(train[f])
    test[f] = indexer.get_indexer(test[f])
for df in [train, test]:
    df['hits/pageviews'] = (df["totals.pageviews"]/(df["totals.hits"])).apply(lambda x: 0 if np.isinf(x) else x)
    df['is_high_hits'] = np.logical_or(df["totals.hits"]>5,df["totals.pageviews"]>5).astype(np.int32)
    df["Revenue"] = np.log1p(df["Revenue"])
    
folds = get_folds(df=train, n_splits=5)
y_reg = y_reg.astype(float)
train_features = [_f for _f in train.columns if _f not in excluded_features]
```

```python
# LGBMRegressor at Session level
train_session = np.zeros(train.shape[0])
test_session = np.zeros(test.shape[0])
for fold_, (trn_, val_) in enumerate(folds):
    trn_x, trn_y = train[train_features].iloc[trn_], y_reg.iloc[trn_]
    val_x, val_y = train[train_features].iloc[val_], y_reg.iloc[val_]   
    reg = lgb.LGBMRegressor(
        num_leaves=31,
        learning_rate=0.03,
        n_estimators=5000,
        subsample=.9,
        colsample_bytree=.9,
        random_state=1
    )
    reg.fit(
        trn_x, np.log1p(trn_y),
        eval_set=[(val_x, np.log1p(val_y))],
        early_stopping_rounds=200,
        verbose=100,
        eval_metric='rmse'
    )
    train_session[val_] = reg.predict(val_x, num_iteration=reg.best_iteration_)
    train_session[train_session < 0] = 0
    _preds = reg.predict(test[train_features], num_iteration=reg.best_iteration_)
    _preds[_preds < 0] = 0
    test_session += np.expm1(_preds) / len(folds)
    
print(mean_squared_error(np.log1p(y_reg), train_session) ** .5)
train['predictions'] = np.expm1(train_session)
test['predictions'] = test_session
```

```python
# LGBMRegressor at Visitor level
def getData(df1):
    data = df1[train_features + ['fullVisitorId']].groupby('fullVisitorId').mean()
    data_pred_list = df1[['fullVisitorId', 'predictions']].groupby('fullVisitorId')\
    .apply(lambda df: list(df.predictions))\
    .apply(lambda x: {'pred_'+str(i): pred for i, pred in enumerate(x)})
    data_pred = pd.DataFrame(list(data_pred_list.values), index=data.index)
    cols = data_pred.columns
    data_pred['t_mean'] = np.log1p(data_pred[cols].mean(axis=1))
    data_pred['t_median'] = np.log1p(data_pred[cols].median(axis=1))
    data_pred['t_sum_log'] = np.log1p(data_pred[cols]).sum(axis=1)
    data_pred['t_sum_act'] = np.log1p(data_pred[cols].fillna(0).sum(axis=1))
    data_pred['t_nb_sess'] = data_pred[cols].isnull().sum(axis=1)
    full_data = pd.concat([data, data_pred], axis=1)
    return full_data
 
full_data = getData(train)
sub_full_data = getData(test)

# Create target at Visitor level
train['target'] = y_reg
train_user_target = train[['fullVisitorId', 'target']].groupby('fullVisitorId').sum()
folds = get_folds(df=full_data[['totals.pageviews']].reset_index(), n_splits=5)
full_preds = np.zeros(full_data.shape[0])
sub_preds = np.zeros(sub_full_data.shape[0])

for fold_, (trn_, val_) in enumerate(folds):
    trn_x, trn_y = full_data.iloc[trn_], train_user_target['target'].iloc[trn_]
    val_x, val_y = full_data.iloc[val_], train_user_target['target'].iloc[val_]
    
    reg = lgb.LGBMRegressor(
        num_leaves=31,
        learning_rate=0.03,
        n_estimators=2000,
        subsample=.9,
        colsample_bytree=.9,
        random_state=1
    )
    reg.fit(
        trn_x, np.log1p(trn_y),
        eval_set=[(trn_x, np.log1p(trn_y)), (val_x, np.log1p(val_y))],
        eval_names=['TRAIN', 'VALID'],
        early_stopping_rounds=200,
        eval_metric='rmse',
        verbose=100
    )
    
    full_preds[val_] = reg.predict(val_x, num_iteration=reg.best_iteration_)
    full_preds[full_preds < 0] = 0
    
    # Make sure features are in the same order
    _preds = reg.predict(sub_full_data[full_data.columns], num_iteration=reg.best_iteration_)
    _preds[_preds < 0] = 0
    sub_preds += _preds / len(folds)
    
print(mean_squared_error(np.log1p(train_user_target['target']), full_preds) ** .5)
```

```python
# submit predictions 
sub_full_data['PredictedLogRevenue'] = sub_preds
sub_full_data[['PredictedLogRevenue']].to_csv('submission.csv', index=True)
```

```python
def col_by_col_count(df, col1, col2, threshold=0):
    return df.groupby([col1]).count()[col2].apply(lambda x: (x if x>threshold else np.nan)).dropna()

def col_by_col_sum(df, col1, col2, threshold=0):
    return df.groupby([col1]).sum(numeric_only=True)[col2].apply(lambda x: (x if x>threshold else np.nan)).dropna()

def myautopct(pct):
    return ('%.2f' % pct) if pct > 2 else ''

def mypie(df, title, angle=0):
    # autopct='%1.1f%%'
    # textprops={'size': 'small'}  (Kaggle python (3.6.6) + libs didn't recognise this)
    df.plot(kind='pie', figsize=(5, 5), radius=1.2, startangle=angle, autopct=myautopct, pctdistance=0.8,
        rotatelabels=False, legend=True, explode=[0.02]*df.size);
    plt.title(title, weight='bold', size=14, x=2.0, y=-0.01);
    plt.xlabel('')
    plt.ylabel('')
    plt.legend(bbox_to_anchor=(2.5, 1.0), ncol=2, fontsize=10, fancybox=True, shadow=True);
```
